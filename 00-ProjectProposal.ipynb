## Research Question

How does sleep duration and regularity relate to academic performance among college students, and can sleep patterns predict GPA?

**Variables:**
- **Dependent**: Cumulative GPA
- **Independent**: Average sleep duration in hours per night, sleep regularity in standard deviation of sleep times
- **Confounding variables**: Academic year, course taken, credit load



## Background and Prior Work

Sleep is an important part of the day for the human body to work. Most people need on average 7-9 hours per day to function at a high level. Sleep deprivation leads to impaired cognitive functions which can lead to a reduction in learning capacity and increase academic errors. Research has shown that students that sleep less than 6 hours a night have significantly lower GPAs compared to students who sleep 7-9 hours. Moreover, research has also shown that sleep consistency is just as important as total duration per night. This means that it is equally important to sleep the recommended hours per night as well as keep it consistent throughout the week. 


### References

- https://doi.org/10.1016/j.sleh.2014.12.010
- https://doi.org/10.1111/j.1365-2869.2010.00856.x
- https://www.pnas.org/doi/10.1073/pnas.2209123120

## Hypothesis






If a student has a consistent sleep schedule of 8 hours or more a day then they are more likely to have a 4.0 than students who sleep less than 6 hours a day

sleep is one of many factors affecting GPA; other factors (motivation, course difficulty, preparation, resources) account for the remaining variance. Most humans require 8-10 hours of sleep a day to perform at a high level so there should be a correlation between sleep and academic performance.
