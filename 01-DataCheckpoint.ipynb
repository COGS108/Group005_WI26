## Background and Prior Work (updated from 00)

### Sleep and Cognitive Function

Sleep plays a fundamental role in maintaining cognitive health and academic performance. As sleep neuroscience research has established, "sleep is essential for memory consolidation, attention restoration, and emotional regulation" (Walker & Stickgold, 2006). The human body requires approximately 7-9 hours of sleep per night to function optimally, a window recommended by sleep medicine specialists based on neuroscientific findings. During sleep, the brain consolidates memories, processes emotional information, and restores metabolic resources necessary for wakeful cognitive performance (Goel, 2017).

Sleep deprivation has profound cognitive consequences. Research demonstrates that when individuals are sleep deprived, there are measurable declines in attention, working memory, and critical abilities for academic success. A meta analysis found that "sleep-deprived individuals showed consistent impairments across multiple cognitive domains, with deficits equivalent to or exceeding those observed with alcohol intoxication" (Lim & Dinges, 2010). For students specifically, this translates to reduced learning capacity, difficulty concentrating during lectures, and increased likelihood of academic errors on exams and assignments.

### Sleep Duration and Academic Performance

Research demonstrates a strong relationship between sleep duration and academic outcomes, particularly in college student populations. Multiple studies have found that college students who sleep fewer than 6 hours per night have significantly lower GPAs compared to their peers who maintain 7-9 hours of sleep nightly. In one study examining 148 college students, researchers found that "students sleeping less than 7 hours per night were significantly more likely to have below-average academic performance," with sleep duration accounting for meaningful variance in GPA (Giri et al., 2021). 

The academic consequences show a linear decline. As sleep duration decreases, academic performance tends to decline correspondingly. This finding is particularly relevant to college populations, where sleep deprivation is often regarded as normal and even celebrated as a marker of dedication to studies. However, this cultural attitude toward sleep deprivation contrasts  with the evidence that sleep is not a luxury but a biological necessity for optimal cognitive performance.

### Sleep Consistency and Regularity

Beyond total sleep duration, research increasingly emphasizes the importance of sleep consistency and regularity. As research has shown, "maintaining a regular sleep schedule is just as important for cognitive function as achieving adequate total sleep hours." This is because   regular sleep patterns help synchronize circadian rhythms with behavioral patterns (Taub & Berger, 1973). Sleep irregularities such as varying bedtimes and wake times across days can disrupt these biological rhythms and compromise the restorative benefits of sleep, even when total duration is adequate.

Studies on circadian misalignment demonstrate that "irregular sleep schedules lead to desynchronization between the internal biological clock and external time cues, resulting in impaired cognitive performance despite nominally adequate sleep duration" (Monk & Folkard). This distinction between sleep quantity and sleep consistency is particularly important for studying college student populations. Students often face irregular schedules due to varying class times, midterms/finals periods, and social commitments. A student sleeping 9 hours but with highly variable sleep times may experience cognitive performance deficits similar to a student sleeping a consistent 7 hours nightly.

### Research Gap

While existing literature establishes the relationship between sleep patterns (both duration and consistency) and academic performance, most studies rely on self-reported data or limited sample sizes. Few studies simultaneously examine both sleep duration and sleep consistency as predictors of academic performance. This project aims to contribute to this research gap by analyzing datasets linking sleep patterns to academic records, while simultaneously examining how sleep consistencyoften overlooked in favor of duration alonepredicts GPA. Understanding both dimensions of sleep behavior may provide more nuanced insights into how sleep affects academic success and could inform evidence-based recommendations for student health interventions.

### Dataset 1: CMU Sleep Study - Nightly Sleep Time and GPA in First-Year College Students

This is a multi-site empirical study examining the relationship between sleep patterns tracked via Fitbit wearables and academic performance in first-year college students. The data comes from five separate cohorts at three universities (CMU, University of Washington, and Notre Dame) studied across multiple years (2016-2019).

**Key Metrics:**
- **Total Sleep Time**: Average actual sleep time per night in minutes (continuous variable, typically 300-600 minutes / 5-10 hours). This is calculated from Fitbit data as the time spent in bed minus awake/restless periods. The recommended range is 420-540 minutes (7-9 hours) for college students.
- **Bedtime Mean Successive Squared Difference (MSSD)**: Measured in hours, this captures day-to-day variability in bedtime (when students go to sleep). Lower values indicate more regular sleep schedules; high variability disrupts circadian rhythms even with adequate total sleep.
- **Midpoint of Sleep**: Average midpoint between bedtime and wake time (in minutes after 11 pm). For example, 364 minutes means approximately 5:04 am. This helps characterize whether students are sleeping during typical night hours or shifted schedules.
- **Term GPA**: End-of-semester cumulative GPA on 0-4.0 scale. This is the primary outcome variable we are predicting.
- **Cumulative GPA (cum_gpa)**: GPA from previous semesters (typically just fall semester for first-years at 3 universities, fall+winter for UW quarters). Important control variable.
- **Term Units (Z-normalized)**: Number of course units/credits taken, normalized as z-scores within each university cohort since universities use different unit systems.
- **Demographics**: Gender (0=male, 1=female), Race (0=underrepresented minority, 1=non-underrepresented), First-gen status (0=non-first-gen, 1=first-gen). Allows testing for fairness and bias.

**Data Collection Methodology:**
- Students carried Fitbits for one month early in their spring academic term
- Sleep episodes identified using specific algorithm: episodes >=20 minutes separated by >=5 minutes
- "Main" sleep episode (longest between noon-to-noon) used as primary sleep measurement
- Grades obtained from university registrars
- 634 total participants across 5 study cohorts

**Data Strengths:**
- Objective, device-measured sleep data (Fitbit actigraphy), eliminating self-report bias
- Published peer-reviewed study with rigorous methodology (Creswell et al., 2023, PNAS)
- Multiple independent universities and time periods increase generalizability
- Large sample size providing statistical power
- Includes both sleep *duration* and sleep *variability* measures aligned with research questions
- Rich demographic data enables fairness and bias analysis

**Data Limitations/Concerns:**
- Limited to first-year college students only (may not generalize to upper-class students or non-traditional students)
- Temporal scope limited to one month early in spring semester (may not capture full semester patterns or seasonal effects)
- Some device non-compliance: users may not have worn Fitbit constantly (frac_nights_with_data variable available as quality indicator)
- Cumulative GPA from *previous* semesters, not concurrent with sleep tracking (introduces timing mismatch)
- May have survivor bias if students who dropped out are underrepresented
- Active tracking during study period may alter behavior (Hawthorne effect)


```python
import pandas as pd
import numpy as np
import os
import urllib.request

print("\n" + "="*80)
print("DATASET 1: CMU Sleep Study - Loading and Cleaning")
print("="*80)

# Create directories and download dataset
os.makedirs('data/00-raw', exist_ok=True)
os.makedirs('data/01-interim', exist_ok=True)
os.makedirs('data/02-processed', exist_ok=True)

raw_path = 'data/00-raw/cmu-sleep.csv'
if not os.path.exists(raw_path):
    url = 'https://cmustatistics.github.io/data-repository/data/cmu-sleep.csv'
    try:
        urllib.request.urlretrieve(url, raw_path)
        print(f"Downloaded CMU dataset to {raw_path}")
    except Exception as e:
        print(f"Error: {e}. Please download from {url} and save to {raw_path}")
else:
    print(f"Dataset already present at {raw_path}")

# Load and inspect
df_cmu = pd.read_csv(raw_path)

print(f"\n--- DATASET SIZE ---")
print(f"Shape: {df_cmu.shape}")
print(f"Columns ({len(df_cmu.columns)}): {df_cmu.columns.tolist()}")

print(f"\n--- DATASET STRUCTURE (TIDINESS) ---")
print(f"Structure: One row per student (observation)")
print(f"Variables: Columns represent sleep metrics, demographics, academic performance")
print(f"Tidy assessment: Yes. Wide format with one observation per row, one variable per column")
print(f"First 3 rows:")
print(df_cmu.head(3))
print(f"\nData types:")
print(df_cmu.dtypes)

print(f"\n--- MISSING DATA ANALYSIS ---")
missing_data = pd.DataFrame({
    'Missing_Count': df_cmu.isnull().sum(),
    'Percentage': (df_cmu.isnull().sum() / len(df_cmu) * 100).round(2),
    'Data_Type': df_cmu.dtypes
})
missing_data_sorted = missing_data[missing_data['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)
if len(missing_data_sorted) == 0:
    print(f"No missing values detected in raw dataset")
else:
    print(missing_data_sorted)
    print(f"\nMissingness pattern: Check if systematic (e.g., all missing in same records)")
    missing_by_record = df_cmu.isnull().sum(axis=1)
    print(f"  Records with 0 missing values: {(missing_by_record == 0).sum()}")
    print(f"  Records with 1+ missing values: {(missing_by_record > 0).sum()}")
    if (missing_by_record > 0).sum() > 0:
        print(f"  Max missing per record: {missing_by_record.max()}")

print(f"\n--- OUTLIER AND SUSPICIOUS VALUE DETECTION ---")
# Check key variables for outliers
key_vars = ['TotalSleepTime', 'term_gpa', 'bedtime_mssd']
if 'frac_nights_with_data' in df_cmu.columns:
    key_vars.append('frac_nights_with_data')

for var in key_vars:
    if var in df_cmu.columns:
        print(f"\n{var}:")
        print(f"  Range: [{df_cmu[var].min():.2f}, {df_cmu[var].max():.2f}]")
        print(f"  Mean: {df_cmu[var].mean():.2f}, Std: {df_cmu[var].std():.2f}")
        print(f"  Q1: {df_cmu[var].quantile(0.25):.2f}, Median: {df_cmu[var].median():.2f}, Q3: {df_cmu[var].quantile(0.75):.2f}")
        
        # Flag outliers (values outside 1.5*IQR)
        Q1 = df_cmu[var].quantile(0.25)
        Q3 = df_cmu[var].quantile(0.75)
        IQR = Q3 - Q1
        outliers = df_cmu[(df_cmu[var] < Q1 - 1.5*IQR) | (df_cmu[var] > Q3 + 1.5*IQR)][var].count()
        print(f"  Potential outliers (IQR method): {outliers} records")
        
        # Logical bounds check
        if var == 'TotalSleepTime':
            impossible = ((df_cmu[var] < 0) | (df_cmu[var] > 1440)).sum()
            print(f"  Impossible values (< 0 or > 1440 min): {impossible}")
        elif var == 'term_gpa':
            impossible = ((df_cmu[var] < 0) | (df_cmu[var] > 4.0)).sum()
            print(f"  Out of bounds values (< 0 or > 4.0): {impossible}")
        elif var == 'frac_nights_with_data':
            impossible = ((df_cmu[var] < 0) | (df_cmu[var] > 1)).sum()
            print(f"  Out of bounds values (< 0 or > 1): {impossible}")

print(f"\n--- DATA CLEANING STRATEGY ---")
print(f"Decision: 1. Remove low-quality records (frac_nights_with_data < 0.60)")
print(f"Justification: Device data with <60% nights tracked is unreliable")
print(f"\nDecision: 2. Remove records missing critical variables (TotalSleepTime, term_gpa, bedtime_mssd)")
print(f"Justification: These are core analysis variables; cannot impute sleep/academic data")
print(f"\nDecision: 3. Remove logical outliers (impossible values)")
print(f"Justification: Values outside biological/measurement bounds indicate data errors")
print(f"\nDecision: 4. Convert TotalSleepTime to hours for interpretability")
print(f"Justification: Hours are more interpretable than minutes for sleep research")

# Data cleaning
df_cmu_clean = df_cmu.copy()

print(f"\n--- CLEANING EXECUTION ---")
print(f"Starting records: {len(df_cmu_clean)}")

# Remove low-quality records
if 'frac_nights_with_data' in df_cmu_clean.columns:
    before = len(df_cmu_clean)
    df_cmu_clean = df_cmu_clean[df_cmu_clean['frac_nights_with_data'] >= 0.60]
    removed = before - len(df_cmu_clean)
    print(f"After quality filter: {len(df_cmu_clean)} records ({removed} removed)")

# Remove missing critical variables
before = len(df_cmu_clean)
df_cmu_clean = df_cmu_clean.dropna(subset=['TotalSleepTime', 'term_gpa', 'bedtime_mssd'])
removed = before - len(df_cmu_clean)
print(f"After missing value removal: {len(df_cmu_clean)} records ({removed} removed)")

# Remove logical outliers
before = len(df_cmu_clean)
df_cmu_clean = df_cmu_clean[
    (df_cmu_clean['TotalSleepTime'] >= 180) & (df_cmu_clean['TotalSleepTime'] <= 1000) &
    (df_cmu_clean['term_gpa'] >= 0) & (df_cmu_clean['term_gpa'] <= 4.0) &
    (df_cmu_clean['bedtime_mssd'] >= 0)
]
removed = before - len(df_cmu_clean)
print(f"After outlier removal: {len(df_cmu_clean)} records ({removed} removed)")

# Create processed variable
df_cmu_clean['TotalSleepTime_hours'] = df_cmu_clean['TotalSleepTime'] / 60
print(f"\nData retention: {len(df_cmu_clean)/len(df_cmu)*100:.1f}% of original records")

print(f"\n--- FINAL DATASET SHAPE ---")
print(f"Shape: {df_cmu_clean.shape}")
print(f"Records: {len(df_cmu_clean)}")
print(f"Variables: {len(df_cmu_clean.columns)}")
print(f"No missing values remaining: {df_cmu_clean.isnull().sum().sum() == 0}")

print(f"\n--- SUMMARY STATISTICS FOR KEY VARIABLES ---")
key_summary_vars = ['TotalSleepTime_hours', 'term_gpa', 'bedtime_mssd']
if 'cum_gpa' in df_cmu_clean.columns:
    key_summary_vars.append('cum_gpa')
print(df_cmu_clean[key_summary_vars].describe())

# Save cleaned data
output_path = 'data/02-processed/cmu_sleep_processed.csv'
df_cmu_clean.to_csv(output_path, index=False)
print(f"\nCleaned data saved to {output_path}")
```

### Integration Strategy: CMU and Kaggle Datasets

These two datasets approach the sleep-performance question from different angles and should be analyzed with different purposes:

**Primary Analysis (CMU Dataset):**
The CMU Sleep Study is the focus of the primary analysis because:
- Objective Fitbit-measured sleep data eliminates recall bias
- Multi-site design with multiple universities and time periods
- Large sample size (634) provides statistical power
- Variables directly measure what we want: sleep *duration* AND sleep *consistency*
- Actual academic performance (GPA) available from university registrars
- Published peer-reviewed methodology ensures data quality

**Supplementary Analysis (Kaggle Dataset):**
The Kaggle Sleep Health and Lifestyle dataset serves to:
- **Validate findings**: Test whether CMU results hold in a different population
- **Explore generalization**: Examine sleep-health relationships across broader age range and occupations
- **Identify confounds**: Use stress, activity, and health variables to explore potential mechanisms
- **Robustness checks**: If sleep-related findings are consistent across these distinct datasets, confidence in conclusions increases

**Data Combination Approach:**
Given the different purposes and data characteristics:

1. **Analyze separately first**: Conduct full analysis on CMU data independently. Kaggle analysis explores sleep-health relationships (not academic performance).

2. **Standardization for comparison** (if combining):
   - Sleep Duration: Convert CMU minutes to hours for direct comparison with Kaggle hours
   - Sleep Quality: CMU's bedtime MSSD (variability) vs. Kaggle's subjective quality rating represent different aspects
   - Demographics: Both have gender/age; ensure consistent coding (0/1 vs. M/F)

3. **Note differences in metadata**:
   - CMU: First-year college students, spring semester, month-long tracking
   - Kaggle: General population spanning multiple ages/occupations, synthetic data

4. **Quality assurance**:
   - Verify Kaggle data ranges are sensible (sleep hours 0-24, quality/stress scales 1-10)
   - Check for impossible values or data entry errors
   - Handle missing values transparently and document decisions
   - Document any data transformations or re-codings

5. **Reporting**:
   - Primary results from CMU data in main narrative
   - Kaggle analysis as validation/supplementary findings
   - Clearly label which dataset each finding comes from

```python
import pandas as pd
import numpy as np

print("\n" + "="*80)
print("DATASET 2: Kaggle Sleep Health and Lifestyle - Loading and Cleaning")
print("="*80)

# Load Kaggle Sleep Health and Lifestyle Dataset
kaggle_path = 'data/00-raw/Sleep_health_and_lifestyle_dataset.csv'
if not os.path.exists(kaggle_path):
    print(f"\nNote: Kaggle dataset must be downloaded manually from:")
    print(f"   https://www.kaggle.com/datasets/uom190346a/sleep-health-and-lifestyle-dataset")
    print(f"   Save as: {kaggle_path}")
    print(f"\nDataset not found. Skipping Kaggle processing...\n")
else:
    df_kaggle = pd.read_csv(kaggle_path)
    
    print(f"\n--- DATASET SIZE ---")
    print(f"Shape: {df_kaggle.shape}")
    print(f"Rows: {len(df_kaggle)}, Columns: {len(df_kaggle.columns)}")
    print(f"Column names: {df_kaggle.columns.tolist()}")
    
    print(f"\n--- DATASET STRUCTURE (TIDINESS) ---")
    print(f"Structure: One row per individual (observation)")
    print(f"Variables: Columns represent sleep quality, health metrics, and demographics")
    print(f"Tidy assessment: Yes. Wide format with one observation per row, one variable per column")
    print(f"First 3 rows:")
    print(df_kaggle.head(3))
    print(f"\nData types:")
    print(df_kaggle.dtypes)
    
    print(f"\n--- MISSING DATA ANALYSIS ---")
    missing_data = pd.DataFrame({
        'Missing_Count': df_kaggle.isnull().sum(),
        'Percentage': (df_kaggle.isnull().sum() / len(df_kaggle) * 100).round(2)
    })
    missing_data_sorted = missing_data[missing_data['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)
    if len(missing_data_sorted) == 0:
        print(f"No missing values detected in raw dataset")
    else:
        print(missing_data_sorted)
        print(f"\nMissingness pattern analysis:")
        missing_by_record = df_kaggle.isnull().sum(axis=1)
        print(f"  Records with 0 missing values: {(missing_by_record == 0).sum()}")
        print(f"  Records with 1+ missing values: {(missing_by_record > 0).sum()}")
    
    print(f"\n--- OUTLIER AND SUSPICIOUS VALUE DETECTION ---")
    # Identify numeric columns
    numeric_cols = df_kaggle.select_dtypes(include=[np.number]).columns.tolist()
    key_numeric_vars = ['Sleep Duration', 'Quality of Sleep', 'Stress Level', 'Heart Rate', 'Daily Steps']
    key_numeric_vars = [col for col in key_numeric_vars if col in numeric_cols]
    
    for var in key_numeric_vars:
        print(f"\n{var}:")
        print(f"  Range: [{df_kaggle[var].min():.2f}, {df_kaggle[var].max():.2f}]")
        print(f"  Mean: {df_kaggle[var].mean():.2f}, Std: {df_kaggle[var].std():.2f}")
        print(f"  Q1: {df_kaggle[var].quantile(0.25):.2f}, Median: {df_kaggle[var].median():.2f}, Q3: {df_kaggle[var].quantile(0.75):.2f}")
        
        # Flag outliers (values outside 1.5*IQR)
        Q1 = df_kaggle[var].quantile(0.25)
        Q3 = df_kaggle[var].quantile(0.75)
        IQR = Q3 - Q1
        outliers = df_kaggle[(df_kaggle[var] < Q1 - 1.5*IQR) | (df_kaggle[var] > Q3 + 1.5*IQR)][var].count()
        print(f"  Potential outliers (IQR method): {outliers} records")
        
        # Logical bounds check
        if 'Sleep Duration' in var:
            impossible = ((df_kaggle[var] < 0) | (df_kaggle[var] > 24)).sum()
            print(f"  Out of bounds (< 0 or > 24 hrs): {impossible}")
        elif 'Quality' in var or 'Stress' in var:
            impossible = ((df_kaggle[var] < 1) | (df_kaggle[var] > 10)).sum()
            print(f"  Out of bounds (< 1 or > 10): {impossible}")
    
    print(f"\n--- CATEGORICAL VARIABLES ---")
    categorical_cols = df_kaggle.select_dtypes(include=['object']).columns.tolist()
    for col in categorical_cols:
        print(f"\n{col}: {df_kaggle[col].nunique()} unique values")
        print(f"  Values: {df_kaggle[col].unique().tolist()}")
        print(f"  Distribution: {df_kaggle[col].value_counts().to_dict()}")
    
    print(f"\n--- DATA CLEANING STRATEGY ---")
    print(f"Decision: 1. Remove records with any missing values (dropna)")
    print(f"Justification: Dataset is small (400 records), missing data is systematic (all in one var)")
    print(f"             Complete cases preserve data integrity and relationships")
    print(f"\nDecision: 2. Validate numeric ranges (logical bounds checking)")
    print(f"Justification: Sleep Duration must be 0-24 hrs, Quality/Stress must be 1-10")
    print(f"             Out-of-bounds values indicate data entry errors")
    print(f"\nDecision: 3. Parse Blood Pressure into systolic/diastolic components")
    print(f"Justification: Separate values needed for individual analysis")
    print(f"\nDecision: 4. Verify categorical variables have expected values")
    print(f"Justification: Identifies typos or unexpected categories")
    
    # ===== DATA CLEANING =====
    print(f"\n--- CLEANING EXECUTION ---")
    print(f"Starting records: {len(df_kaggle)}")
    
    df_kaggle_clean = df_kaggle.copy()
    
    # Step 1: Remove rows with missing values
    initial_rows = len(df_kaggle_clean)
    df_kaggle_clean = df_kaggle_clean.dropna()
    removed = initial_rows - len(df_kaggle_clean)
    print(f"After missing value removal: {len(df_kaggle_clean)} records ({removed} removed)")
    print(f"Data retention so far: {len(df_kaggle_clean)/len(df_kaggle)*100:.1f}%")
    
    # Step 2: Validate ranges for sleep-related variables
    initial_rows = len(df_kaggle_clean)
    
    if 'Sleep Duration' in df_kaggle_clean.columns:
        df_kaggle_clean = df_kaggle_clean[
            (df_kaggle_clean['Sleep Duration'] > 0) & (df_kaggle_clean['Sleep Duration'] <= 24)
        ]
    
    if 'Quality of Sleep' in df_kaggle_clean.columns:
        df_kaggle_clean = df_kaggle_clean[
            (df_kaggle_clean['Quality of Sleep'] >= 1) & (df_kaggle_clean['Quality of Sleep'] <= 10)
        ]
    
    if 'Stress Level' in df_kaggle_clean.columns:
        df_kaggle_clean = df_kaggle_clean[
            (df_kaggle_clean['Stress Level'] >= 1) & (df_kaggle_clean['Stress Level'] <= 10)
        ]
    
    if 'Daily Steps' in df_kaggle_clean.columns:
        df_kaggle_clean = df_kaggle_clean[(df_kaggle_clean['Daily Steps'] > 0)]
    
    removed = initial_rows - len(df_kaggle_clean)
    if removed > 0:
        print(f"After range validation: {len(df_kaggle_clean)} records ({removed} removed)")
    else:
        print(f"After range validation: All records within expected ranges")
    
    # Step 3: Parse Blood Pressure if needed (separate systolic/diastolic)
    if 'Blood Pressure' in df_kaggle_clean.columns:
        print(f"\nProcessing Blood Pressure column...")
        try:
            # Split Blood Pressure into systolic and diastolic
            bp_split = df_kaggle_clean['Blood Pressure'].str.split('/', expand=True)
            df_kaggle_clean['Systolic_BP'] = pd.to_numeric(bp_split[0].str.strip(), errors='coerce')
            df_kaggle_clean['Diastolic_BP'] = pd.to_numeric(bp_split[1].str.strip(), errors='coerce')
            print(f"Blood Pressure parsed into Systolic_BP and Diastolic_BP")
            print(f"  Systolic range: [{df_kaggle_clean['Systolic_BP'].min():.0f}, {df_kaggle_clean['Systolic_BP'].max():.0f}]")
            print(f"  Diastolic range: [{df_kaggle_clean['Diastolic_BP'].min():.0f}, {df_kaggle_clean['Diastolic_BP'].max():.0f}]")
        except Exception as e:
            print(f"Could not parse Blood Pressure: {e}")
    
    # Step 4: Verify categorical variables
    print(f"\n--- CATEGORICAL VARIABLE VERIFICATION ---")
    if 'Gender' in df_kaggle_clean.columns:
        print(f"Gender unique values: {df_kaggle_clean['Gender'].unique().tolist()}")
    if 'BMI Category' in df_kaggle_clean.columns:
        print(f"BMI Category unique values: {df_kaggle_clean['BMI Category'].unique().tolist()}")
    if 'Sleep Disorder' in df_kaggle_clean.columns:
        print(f"Sleep Disorder unique values: {df_kaggle_clean['Sleep Disorder'].unique().tolist()}")
        print(f"Sleep Disorder distribution:")
        print(df_kaggle_clean['Sleep Disorder'].value_counts())
    
    print(f"\n--- FINAL DATASET ---")
    print(f"Shape: {df_kaggle_clean.shape}")
    print(f"Records: {len(df_kaggle_clean)}")
    print(f"Variables: {len(df_kaggle_clean.columns)}")
    print(f"Data retention: {len(df_kaggle_clean)/len(df_kaggle)*100:.1f}% of original records")
    print(f"No missing values remaining: {df_kaggle_clean.isnull().sum().sum() == 0}")
    
    print(f"\n--- SUMMARY STATISTICS FOR KEY VARIABLES ---")
    key_summary_vars = ['Sleep Duration', 'Quality of Sleep', 'Stress Level', 'Heart Rate', 'Daily Steps']
    key_summary_vars = [col for col in key_summary_vars if col in df_kaggle_clean.columns]
    if key_summary_vars:
        print(df_kaggle_clean[key_summary_vars].describe())
    
    # Step 5: Save cleaned data
    output_path = 'data/02-processed/kaggle_sleep_health_processed.csv'
    df_kaggle_clean.to_csv(output_path, index=False)
    print(f"\nCleaned data saved to {output_path}")
```
